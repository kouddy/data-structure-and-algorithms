```python
def swap(arr, i, j):
  tmp = arr[i]
  arr[i] = arr[j]
  arr[j] = tmp

def bubble_sort(arr):
  for i in range(len(arr)-1):
    for j in range(i+1, len(arr)):
      if arr[i] > arr[j]: swap(arr, i, j)

def selection_sort(arr):
  for i in range(len(arr)):
    min_index = i
    for j in range(i, len(arr)):
      if arr[j] < arr[min_index]: min_index = j
    swap(arr, i, min_index)

def merge(arr1, arr2):
  result = []
  i1 = i2 = 0
  while i1 < len(arr1) and i2 < len(arr2):
    if arr1[i1] < arr2[i2]:
      result.append(arr1[i1])
      i1 += 1
    else:
      result.append(arr2[i2])
      i2 += 1

  result.extend(arr1[i1:])
  result.extend(arr2[i2:])
  return result

def merge_sort(arr): 
  if len(arr) <= 1: return arr
  mid = int(len(arr) / 2)
  left = merge_sort(arr[0:mid])
  right = merge_sort(arr[mid:])
  return merge(left, right)

def quick_sort(arr):
  def _quick_sort(arr, low, highExc):
    if low < highExc:
      p = partition(arr, low, highExc)
      _quick_sort(arr, low, p)
      _quick_sort(arr, p + 1, highExc)

  _quick_sort(arr, 0, len(arr))

def partition(arr, low, highExc):
  if low < highExc:
    p = arr[highExc-1]
    low_i = low
    for i in range(low, highExc):
      if arr[i] <= p:
        swap(arr, i, low_i)
        low_i += 1
  return low_i - 1
```

```python
arr = [1, 4, 3, 9, 8, 0]
bubble_sort(arr)
print(arr)

arr = [1, 4, 3, 9, 8, 0]
selection_sort(arr)
print(arr)

arr = [1, 4, 3, 9, 8, 0]
arr = merge_sort(arr)
print(arr)

arr = [1, 4, 3, 9, 8, 0]
quick_sort(arr)
print(arr)
```
